```python
#nan값 있는 컬럼
isna_list=[]
for i in x:
    try:
        x[i].isna().value_counts()[True]
        isna_list.append(i)
        # print(i)
    except:
        continue
```

```python
#nan 비율까지 
isna_ratio =[]
for i in x[isna_list]:
    isna_ratio.append([i,round(x[i].isna().value_counts()[True]/len(x) *100,3)])
    # print(x[i].isna().value_counts()[True]/len(x) *100)
```

```python
names = [i[0] for i in isna_ratio]
ratio = [i[1] for i in isna_ratio]
```

```python
name_ratio = pd.DataFrame({'ratio':ratio}, index=names)
name_ratio
```

```python
name_ratio.sort_values(by='ratio').plot.barh(figsize=(15,30))
```

```python
def plot_hist(data, columns, nrow, ncol, figsize, hue_value=None):
    # find the distubution of the data. ( visualization would be so good)
    fig, ax = plt.subplots(nrow,ncol, figsize=figsize)
    col, row = ncol,nrow
    col_count = 0
    sns.set_style('dark')
    for r in range(row):
        for c in range(col):
            if col_count >= len(columns):
                ax[r,c].text(0.5, 0.5, "no data")
            else:
                sns.kdeplot(data=data, x=columns[col_count], hue=hue_value, ax=ax[r, c], palette=['#9900cc','#99ff99'],
                                fill = True, hue_order=[1,0], legend = True)
                ax[r,c].set(xlabel = columns[col_count], ylabel=("Density" if c==0 else ''))
                col_count +=1
        # print("col count ", col_count)
```

```python
# Find the distribution of Delinquency variables
d_feats = [c for c in x[D_col] if c not in disc_col]
plot_hist(pd.merge(x,y, left_index=True, right_index=True), d_feats, 15, 6, (50,100),hue_value='target')
```
